```python
#Argument Types:

-->Positional Arguments
-->Default Arguments
-->KeyWord Arguments
-->variable length arguments
-->variable length keyWord arguments
```

```python
#positional

def pos(name,age,items):
    if items["Cname"] == "AI":
         print(name,age)

pos("Anguraj",15, {"Cname":"AI"})
```

```python
#default

def defaultValue(name,age,gender = "male"):
    print(name,age,gender)

defaultValue("Angu",23)
defaultValue("Ishu",23,"Female")
```

```python
#keyword

def keyWord(name,age,items1):
    if items1["Cname"] == "AI":
         print(name,age)

keyWord(name = "Anguraj",age = 15, items1 = {"Cname":"AI"})
```

```python
#variable length


def varLength(name,*marks):
    if sum(marks) > 200:
       result = "Pass"
    else:
        result = "fail"

    print(f" {name} is {result}")

varLength("Anguraj", 7,4,20)
```

```python
#variable length keyw0ard arguments

def varLengthKeyword(name,**items):
    if sum(items["marks"]) > 200 and items["course"] == "AI":
       result = "Pass"
    else:
        result = "fail"

    print(f" {name} is {result}")

varLengthKeyword("Anguraj", course = "AI",marks = [60,75,85,80,50] )

```
