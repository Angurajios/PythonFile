```python
#polymorphism ---- same method name but different behaviour

method overloading --  same methos with diff arguments in same class

method overiding--- same method with same arguments

operator overloading --- same operator but output different
```

```python
import math

class shape:
    def area(self):
        pass


class circle(shape):

    def __init__(self,radius):
        self.radius = radius

    def area(self):

        return math.pi * self.radius * self.radius

class rectangle(shape):
    
     def __init__(self,length,width):
        self.length = length
        self.width = width

     def area(self):
        return 0.5 * self.length * self.width   


shapes = [rectangle(12,18),circle(12)]

for shape  in shapes:
    print(shape.area())
```
