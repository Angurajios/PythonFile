```python
#filter() - filter the given collection by condition or function
#map() -  convert into new collection based on function
#reduce() - convert collection into single value( but import it from functool libraries)
```

```python
#filter str length greater then 5

list1 = ["Anguraj", "mom" , "is", "apple","ishwarya",]

list1 = list(filter(lambda a: len(a)>5,list1))

list1
```

```python
#find the greatest value in the list

list4 = [2,42,34,87,45,98]

from functools import reduce

list4 = reduce(lambda a,b: a if a>b else b, list4)
list4
```

```python
#find the product of list
list3 = [2,3,3,5,4,8]

from functools import reduce

list3 = reduce(lambda a,b: a*b , list3)
list3
```

```python
#square the even numbers in the list
list2 = [2,42,34,87,45,98]

from functools import reduce

list2 = list(map(lambda a: pow(a,2),filter(lambda b: b%2 == 0,list2)))

list2
```
