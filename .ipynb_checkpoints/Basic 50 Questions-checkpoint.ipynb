```python
#14. Write a Python program to find the largest of three numbers.

def largestNumber(n1,n2,n3):
    if n1>n2 and n1>n3:
        print(f"{n1} is Greater")
    elif n2>n3 and n2>n1:
        print(f"{n2} is Greater")
    else:
        print(f"{n3} is Greater")
        

n1 = int(input("Enter number1"))
n2 = int(input("Enter number2"))
n3 = int(input("Enter number3"))

largestNumber(n1,n2,n3)
```

```python
#15. Write a Python program to convert temperatures to and from Celsius and Fahrenheit.
#Formula: c = (f - 32) × (5/9)

tempVale = int(input("Enter the temperature Value"))
unit = input("Enter the unit C/F")

def converTemp(temp,unit):
    if unit.upper() == "F":
        celsius = (temp - 32) * (5/9)
        print(f"Celsius = {celsius}")
    elif unit.upper() == "C":
        Fahrenheit = (temp+32) * (9/5)
        print(f"Fahrenheit = {Fahrenheit}")
    else:
        print("Enter the valid Unit C or F")

    
converTemp(tempVale,unit)
```

```python
#17. Write a program that accepts the lengths of three sides of a triangle as inputs. The program output should indicate whether or not the triangle is a right triangle.

def isRightAngleOrNot(s1,s2,s3):

    sides = [s1,s2,s3]
    largestSide = max(s1,s2,s3)
    sides.remove(largestSide)

    sq_s1 = pow(sides[0],2)
    sq_s2 = pow(sides[1],2)
    sq_s3 = pow(largestSide,2)

    if (sq_s1 + sq_s2)  == sq_s3:
        print("Its a rightAngle triangle")
    else:
        print("Its a not rightAngle triangle")

s1 = int(input())
s2 = int(input())
s3 = int(input())

isRightAngleOrNot(s1,s2,s3)



```

```python
#27.Write a function "perfect()" that determines if a parameter number is a perfect number. Use this function in a program that determines and prints all the perfect numbers between 1 and 1000.


def isPerfect(number):
    perfectNum = 0
    for i in range(1,number):
        if number%i == 0:
            perfectNum+=i
    
    if perfectNum == number:
        return True
    else:
        return False

def find_Perfect_numbers(numberUpto):
    for i in range(1,numberUpto+1):
        if isPerfect(i):
            print(i)



numberUpto = int(input("Enter the Number:"))

find_Perfect_numbers(numberUpto)

```

```python
### 29. Write a function to check if a number is prime or not.

def isPrimeNumber(numb):
    if numb < 2:
        return False
    for i in range(2,numb):
        if numb%i == 0:
            return False
            break
    return True

        
numb = int(input())

if isPrimeNumber(numb):
    print("Its a prime Number")
else:
    print("Its a Not a prime Number")
```

```python
#35. Remove vowels from a string.

def removeVowels(String1):
    vowels = "aeiou"
    for i in String1:        
        if i.lower() in vowels:
            String1 = String1.replace(i,"")
    print(String1)

String1 = input()
removeVowels(String1)
```

```python
#40 Write a program to calculate the electricity bill (accept number of units from user) according to the following criteria:
"""Unit                          Price
First 100 units               No charge
Next 100 units                Rs 5 per unit
After 200 units               Rs 10 per unit

(For example: if input unit is 350, then total bill amount is Rs 2000)"""

def getElectricityBill(units):

    if units <=100:
        print("No charge")
    elif units > 100 and units <=200:
        units -= 100
        price = units * 5
        return(f"The total electricty bill amount is Rs {price}")
    else:
        charge100to200 = 100 * 5              
        chargeAbove200 = (units - 200) * 10   
        price = charge100to200 + chargeAbove200
        return(f"The total electricty bill amount is Rs {price}")



units = int(input("Enter the Units:"))

print(getElectricityBill(units)) 
```

```python
#42. Write a program to check whether a year is a leap year or not.

def findLeapYear(year):
    if year%4 == 0:
        if year%100 ==0:
            if year%400 == 0:
                print("Its a leap year")
            else:
                print("Its a Not leap year")
        else:
            print("Its a leap year")
    else:
        print("Its a Not leap year")  

year = int(input())

findLeapYear(year)
            
```

```python
#43. Write a program to accept a number from 1 to 7 and display the name of the day (e.g., 1 for Sunday, 2 for Monday, and so on).

def findDayOfWeek(day):
    if day>0 and  day <=7:
        days = {1: "Sunday", 2: "Monday", 3: "Tuesday",4: "Wednesday",5: "Thursday",6: "Friday",7: "Satrurday"}
    
        return days[day]
    else:
        return ("Enter the Day 1 to 7")


day = int(input())

findDayOfWeek(day)

```
