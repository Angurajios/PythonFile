```python
#sum of N Natural number

N = int(input())

sum = 0

for i in range(N+1):
    sum += i
print(sum)
```

```python
#reverse Number
number = int(input())

while(number != 0): 
    
    output = number % 10
    print(output,end = "")
    number = number//10

```

```python
#reverse String

Str = input()
output = str()
for i in range(0,len(Str)):
    output = Str[len(Str)-1]
    Str = Str[:len(Str)-1]
    print(output,end ="")



```

```python
#Plaindrom check

Str1 = input()
orginal = Str1
reveresedStr = ""
while Str1:
    reveresedStr += Str1[-1]
    Str1 = Str1[:-1]

print(reveresedStr)
if orginal.lower() == reveresedStr.lower():
    print("Its Palindrome")
else:
    print("Its not")


        
```

```python
##Count the number of digits in a number

digits = int(input())

count = 0
for i in range(0,len(str(digits))):
    count +=1

print(f"given number is {count} digits")
```

```python
#Find the greatest digit in a number

ip = int(input())

highestValue = 0

for i in str(ip):
    if int(i) > highestValue:
        highestValue = int(i)
        
print(f"{highestValue} is greatedVlaue")
```

```python
#find GCD
a = int(input())
b = int(input())

for i in range(1, min(a, b) + 1):
     if a % i == 0 and b % i == 0:
        gcd = i
print(gcd)


```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
lcm = max(a, b)
while True:
    if lcm % a == 0 and lcm % b == 0:
        break
    lcm += 1
print(f"LCM of {a} and {b} is: {lcm}")
```

```python
#Amstrong number

n = int(input("Enter number: "))
total = 0

for i in str(n):
    total += int(i) ** len(str(n))

if n == total:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
numb = int(input())

isPrime = True
for i in range(2,numb):
    if numb % i == 0:
        isPrime = False

print(isPrime)
        
```

```python
numb1 = int(input())

for num in range(1,numb1 +1):
    isPrime = True
    for i in range(2,num):
        if num % i == 0:
            isPrime = False
            break
    if isPrime:
        print(num,end = " ")

```
