```python
#6. Check if a number is a palindrome

number1 = int(input())
original_num = number1
reversed_num = 0  

while number1 != 0:
    digit = number1 % 10        
    reversed_num = reversed_num * 10 + digit  
    number1 = number1 // 10       

if original_num == reversed_num:
    print(f"{original_num} is a Palindrome")
else:
    print(f"{original_num} is not a Palindrome")
```

```python
#18. Find GCD (HCF) of two numbers using loop

a = int(input())
b = int(input())

for i in range(1, min(a, b) + 1):
     if a % i == 0 and b % i == 0:
        gcd = i
print(gcd)
```

```python
#19. Find LCM of two numbers using loop

a = int(input())
b = int(input())

lcm = max(a, b)
while True:
    if lcm % a == 0 and lcm % b == 0:
        break
    lcm += 1
print(f"LCM of {a} and {b} is: {lcm}")
```

```python
#20. Check if a number is an Armstrong number

n = int(input("Enter number: "))
total = 0

for i in str(n):
    total += int(i) ** len(str(n))

if n == total:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
    
```

```python
#21. Print all prime numbers in a given range (1 to N)

numb1 = int(input())

for num in range(1,numb1 +1):
    isPrime = True
    for i in range(2,num):
        if num % i == 0:
            isPrime = False
            break
    if isPrime:
        print(num,end = " ")
        
```

```python
#22. Print all palindrome numbers in a range (1 to N)

numb2 = int(input())

def isPalindrom(number1):
    original_num = number1
    reversed_num = ""
    
    while number1 != 0:
        digit = str(number1 % 10)        
        reversed_num += digit 
        number1 = number1 // 10       


    if original_num == int(reversed_num):
        return True
    else:
        return False



for i in range(11,numb2+1):
    
    if isPalindrom(i):
        print(i)

   
        
```

```python
#25. Count number of vowels and consonants in a string

givenStr = input()

vowels = "aeiou"

vowelCount = 0

for i in vowels:
    if i in givenStr:
        vowelCount+=1

print(f"{vowelCount} vowels in  given String")
print(f"{len(givenStr) - vowelCount}  consonants in  given String")
```

```python
#26. Check if a number is a perfect number

num = int(input())
divisor_sum = 0

for i in range(1, num):
    if num % i == 0:
        divisor_sum += i

if divisor_sum == num:
    print(f"{num} is a Perfect Number")
else:
    print(f"{num} is not a Perfect Number")
```

```python
#27. Check if a string is a palindrome

Str1 = input()
orginal = Str1
reveresedStr = ""
while Str1:
    reveresedStr += Str1[-1]
    Str1 = Str1[:-1]

print(reveresedStr)
if orginal.lower() == reveresedStr.lower():
    print("Its Palindrome")
else:
    print("Its not")
```

```python
#28. Check if a number is a Harshad number (divisible by sum of digits)

number4 = int(input())

sumOfnum4 = 0

for i in str(number4):
    sumOfnum4+=int(i)

print(sumOfnum4)

if number4%sumOfnum4 == 0:
    print(f"{number4} is Harshad number")
else:
    print(f"{number4} is not Harshad number")
```

```python
#29. Fibonacci series up to N terms

number5 = int(input())

a, b = 0, 1

for i in range(number5):
    print(a,end = " ")
    a,b = b,a+b
    
```

```python
#30. Print numbers in words (e.g., 123 → "One Two Three")

number6 = int(input())

wrdStr = ""

dict1 = {
    0: "Zero", 1: "One", 2: "Two", 3: "Three", 4: "Four",
    5: "Five", 6: "Six", 7: "Seven", 8: "Eight", 9: "Nine"
}

for i in str(number6):

    wrdStr += dict1[int(i)] + " "
    
print(wrdStr, end = " ")
```
