```python

#Class and object

class Laptop:
    def __init__(self,brand,price,ram,rom):
        self.brand = brand
        self.price = price
        self.ram = ram
        self.rom = rom
        print("initialized")
        
    def show(self):
        print("8*********")
        print(f"laptop brand {self.brand}")
        print(f"laptop price {self.price}")
        print(f"laptop ram {self.ram}")
        print(f"laptop rom {self.rom}")


l1 = Laptop("Hp",50000,"16gb","1Tb")
l1.show()
        
```
