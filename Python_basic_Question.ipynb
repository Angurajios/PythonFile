```python
#23. Take marks as input and print grade based on conditions (A, B, C, Fail).
marks = int(input("Enter the Marks:"))

if marks >= 90 and marks <= 100:
    print(" A Grade")
elif marks >= 70 and marks <= 89:
    print(" B Grade")
elif marks >= 50 and marks <= 69:
    print(" C Grade")
else:
    print("Fail")

```
