# inheritance---------->one class inherits properties,methods from another class

# why inheritance?
code reusability

types of inheritance:
single inheritance
multiple inheritance
multilevel inheritance
hierchical inheritance
hybrid inheritance

```python
#single inheritance:

class animal:
    def show(self):
        print("Animals...........")

class dog(animal):
    def sound(self):
        print("Dogs makes sound")
    def show(self):
        super().show()
        print("Dogs shows")
        animal.show(self)


d1=dog()
d1.sound()
d1.show()
```

```python
#multilevel inheritance

class grandparent:
    def work(self):
        print(f"Grandpa reading news paper")
class parent(grandparent):
    def display(self,name):
        print(f"parent name {name}")
    def work(self):
        print(f"parents working in a company")
        super().work()

class child(parent):
    def show(self):
        print(f"child.......")
    def work(self):
        print(f"child studying")
        super().work()

c1=child()
c1.display("priya")
c1.work()
```

```python
#multiple

class father:
    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class mother:
    def show(self,name,job):
        print(f"Name:{name}")
        print(f"job:{job}")

class son(father,mother):
    def display(self,name,job):
        print(f"name:{name}")
        print(f"job:{job}")

s1=son()
s1.display("sri","ai engineer")

s1.show("priya","housewi)
```

```python
#multiple

class father:
    def __init__(self,name,job):
        self.name=name
        self.job=job

    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class mother:
    def show(self):
        print(f"Name:{self.name}")
        print(f"job:{self.job}")

class son(father,mother):
    def display(self,name,job):
        print(f"name:{name}")
        print(f"job:{job}")
        print("**********")
        super().display("ram","company")

s1=son("guest","Unemployee")
s1.display("sri","ai engineer")

s1.show()
```

```python
#hierchical inheritance

class parent:
    def show(self):
        print("parents...........")
class child(parent):
    pass

class son(parent):
    pass


s1=son()
c1=child()

s1.show()
c1.show()
```
